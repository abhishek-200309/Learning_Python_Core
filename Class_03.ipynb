# Random Module

- The `random` module is used to generate random values such as integers and floats.
- It can also be used to pick random items from sequence data types.
- We use the `import` keyword to bring a module into Python.

```python
import random as rd
```

## Key Functions of the `random` Module

### `random.random()`
- Generates a random float between `0.0` (inclusive) and `1.0` (exclusive).

```python
# generating random numbers between 0 and 1.
rd.random()
```

```python
# fixing the random numbers generrated using random.seed.

rd.seed(1)
rd.random()
```

```python


rd.random()*10+10
```

---

### `random.uniform()`
- Generates a random floating-point number between two given values.


```python
rd.uniform(1,10)
```

---

### `random.randint()`
- Generates a random integer within the given range.
- The ending value is also included.

```python
rd.randint(1,30)
```

---

### `random.randrange()`
- Generates a random integer from the specified range, similar to `range()`.
- The step value is optional and defaults to `1`.

```python
rd.randrange(0,10,2)
```

---

### `random.choice()`
- Picks a single random value from the given sequence data type.

```python
fruits=['Banana','apple','guava','grapes']

rd.choice(fruits)
```

```python
fruits1=('Banana','apple','guava','grapes')
rd.choice(fruits1)
```

```python
string='abcdefghi'
rd.choice(string)
```

---

### `random.choices()`
- Picks multiple values from a given sequence data type using the `k` parameter.
- Since it samples with replacement, values can repeat in the result.

```python
rd.seed(0)
fruits2=('Banana','mango','pineapple','apple','guava','grapes')
rd.choices(fruits2,k=2)
```

---

### `random.sample()`
- Picks multiple unique items from the given sequence data type using the `k` parameter.

```python
fruits4=['Banana','apple','guava','papaya','pineapple','grapes']
rd.sample(fruits4,k=2)
```

---

### `random.shuffle()`
- Changes the order of items in a sequence randomly.
- The sequence must be mutable, such as a list; otherwise, it will raise an error.

```python
rd.shuffle(fruits4)
```
