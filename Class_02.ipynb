# Packing and Unpacking in Python 

## Packing
- Packing means grouping multiple values into a single variable.
- `x = [1, 2, 3]` stores several values inside one list.
- `a = 1, 2, 3` automatically packs the comma-separated values into a tuple.

```python
x=[12,14,15,16,18,18,17,4,4,5,2,20415,1425] # this is packing storing more than one values in single variable.
```

```python
print(type(x))

x[0]
```

```python
a=1,5,24,47,51,254,14,12  # it will give tuple sequence data type by itself 
```

```python
print(type(a))

a[0]
```

## Unpacking
- Unpacking means extracting values from a sequence and assigning them to separate variables.
- In standard unpacking, the number of variables must exactly match the number of items in the sequence.

```python
a,b,c,d,e,f,g,h,i,j,k,l,m=x
```

```python
display(a,b,c,d,e,f,g,h,i)
```

```python
# this will give error as we didn't gave the enough variable given values must be equal to the given variables vice versa 

a,b,c,d,e,f,g,h,i,j,k,m=x
```

## Extended Unpacking
- Extended unpacking allows one variable to capture all remaining items during assignment.
- We use `*` with a variable, for example `first, *rest = [1, 2, 3, 4]`.
- The starred variable stores the leftover values as a list.
- This is useful when the number of remaining items can vary.

```python
q,w,e,*r,t,y=x


display(q,w,e,t,y)

display(*r) ## this is extended unpacking
```

## Skipping Values While Unpacking
- During unpacking, `_` is commonly used as a placeholder for values you do not plan to use.
- This improves readability when you only need specific elements from a sequence.
- `_` is only a naming convention; Python does not treat it as a special keyword.

```python
m,n,b,v,p,*_=x
```

```python
print(m,n,b,v,p) ## when underscore is used while unpacking it usally means that we don't need that value stored in that variable so we ignore it while working.

print(*_)
```

## Nested Unpacking
- Nested unpacking is used when a sequence contains other sequences inside it.
- It lets you unpack inner values directly by matching the structure of the nested sequence.

```python
x1=[1,2,3,[4,7,8],47,25,[57,54,74],[[[[[4,4,8,]]]]]]
```

```python
print(len(x1))
```

```python
a,b,c,[d,e,f],g,h,[i,j,k],[[[[[l,m,n]]]]],x1
```

```python
display(d,e,f)
```

# User Input Function
- `input()` is a built-in Python function used to take input from the user.
- Syntax: `input('prompt')`, where the prompt is the message shown before the user enters a value.
- The returned value is usually stored in a variable.
- By default, `input()` returns data as a string.
- We can use explicit type casting to convert it to another data type.
- `input()` makes programs more dynamic and interactive.

```python
Name,input('Please Enter Your Name')
age,input('please enter your age')
print('Hi your name is '+Name+' and you are '+age+' years old')
```

```python
## write a code to perform addition of the given numbers

x=float(input('please enter the digit'))

y=float(input('please enter the digit'))


print('The sum of your digits is =' , int(x+y))
```

```python
## write a code to perform subtraction of the given numbers

a=float(input('enter the digits'))
b=float(input('enter the digits'))

print('The result of the subtraction of your given digits is ',int(a-b))
```

### Using `map()` with the `input()` Function
- `map()` can be used to convert the data type of multiple input values in a single step.
- Common syntax: `map(int, input('Enter values separated by commas: ').split(','))`
- This works well when all input values need the same data type.
- A limitation of `map()` is that it applies the same conversion to every value, so it is not suitable for mixed data types.
- In such cases, split the input first and convert each value separately.

```python
a,b,c,d,e,f=map(int,input('enter the digit delemited with ,').split(","))
```
