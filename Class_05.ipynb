# Python Sequence Data-Types
- An ordered collection of elements accessible by its position(index).
- Types of sequence data types `(list,strings,tuples)`

```python
# Example

y='data science'

y[3]
```

```python
a=[1,65,48,'hello']
```

```python
a[3][::-1]
```

## Purpose of Sequence Data-Types
- It can store multiple values at a single place.
- All the values defined while creation has a fixed position `(index)`.
- If we start from left-right the `index` position starts from 0.
- If we start from right-left the `index` position starts from -1.
- Indexing and slicing support.
- `Indexing` means to extract one value from the sequence-data type.
- `Slicing` menas to extract more than one value from the sequence data type.
- Sequence Data-Types are `Iterable` meaning we can fetch there element one by one.
- Sequence Data-Types are `Subscriptable` meaning we can update values using their `index` position





```python
# Indexing
a='hello'

print(a[0])


# Slicing
print(a[0:2])
```

## Types Of Sequence Data-Ypes

### String 

- It represent textual value and `immutable` in nature, meaning we can not changes their values using `index` once created.
- Agar string mai changes karne hai to alag se changes karne honge or fir un changes ko fir se uss variable mai store karna hoga.
### Tuple
- Fixed collection of `heterogeneous` values.
- `Immutable` in nature that makes them suitable for storing permanent values.
- To change the tuple we need to convert it to list and make the changes then agian change it back to tuple.
### List
- It is `heterogeneous` in nature meaning it contains all types of data type.
- It is `mutable` in nature meaning the values present in `list` can be changed in place or removed or appned.
- Changing the same list can be done using `indexing,slicing` or any other list functions.
### Range 
- It is a function which generates an seuence of integer values defined by `start,stop,step`.
- Start is `inclusive` and stop is `exclusive`.
- It can be `mutable or immutable` depending on use case.
- It supports both forward or backward traversal meaning => hum increasing or decreasing order dono mai number generte kar skte hai
- Output print karne ke liye hamey ek constructer ki help leni hoti hai jaise `list,set,tupple, etc.`.

```python
# list
b=[1.4,1,'abhishek',True,3+4j]

# updating list using indexing.

b[0]=51432471812
print(b)

# updating list using slicing.

b[0:2]=[1,2]
print(b)

# adding a new value inside list using append()

b.append('hello')

print(b)
```
