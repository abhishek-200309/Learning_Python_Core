# Python Sequence Data-Types
- An ordered collection of elements accessible by its position(index).
- Types of sequence data types `(list,strings,tuples)`

```python
# Example

y='data science'

y[3]
```

```python
a=[1,65,48,'hello']
```

```python
a[3][::-1]
```

## Purpose of Sequence Data-Types
- It can store multiple values at a single place.
- All the values defined while creation has a fixed position `(index)`.
- If we start from left-right the `index` position starts from 0.
- If we start from right-left the `index` position starts from -1.
- Indexing and slicing support.
- `Indexing` means to extract one value from the sequence-data type.
- `Slicing` menas to extract more than one value from the sequence data type.
- Sequence Data-Types are `Iterable` meaning we can fetch there element one by one.
- Sequence Data-Types are `Subscriptable` meaning we can update values using their `index` position





```python
# Indexing
a='hello'

print(a[0])


# Slicing
print(a[0:2])
```
