### Rules for creating Variables

##### Varibale Naming Structure
- Variable name's first letter be either alpha or underscore(_)
- We can create variables using letters,digits or underscore.
- it can have digits but not in the starting of name of variable
- Special Character is not allowed while creating variable only undersocre (_)is allowed
- Python is Case sensitive
- Can not use python built in keyword for creating variable python will not trow error but that keyword will loose it's identity.

#### Different Variables Naming Style

###### Camel Case Style
- Camel Case Style:- the first word of the variable's name is lower and each new word starts with an uppercase
- not recommended by Python's offical style guide it is more common in (java or c++).
- examples:- myVariableName,studentCount,totalMarksObtained,averageScoreValue,cariableClass.
