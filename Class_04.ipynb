# Python Operators

Learn how Python operators work and practice the arithmetic operators used in expressions.

## What Are Operators?
- Operators are symbols or keywords used to perform actions on values.
- Symbol-based operators include `+`, `-`, `*`, `/`, `%`, `>`, `<`, `!=`, `>=`, `<=`, and `==`.
- Keyword-based operators include `and`, `or`, `not`, `in`, and `not in`.

## Types of Operators
- Arithmetic
- Assignment
- Comparison
- Logical
- Membership
- Identity

### Arithmetic Operators
Arithmetic operators are used to perform mathematical calculations in Python.

#### Addition, Subtraction, Multiplication, and Division
- `+` adds values.
- `-` subtracts values.
- `*` multiplies values.
- `/` divides values and returns a decimal result.

```python
x = 10
a = 15

print(x + a)
print(a - x)
print(x * a)
print(a / x)
```

#### Floor Division (`//`)
- Floor division returns the quotient without the decimal part.
- It is useful when you need a whole-number result.

```python
x = 10
z = 20

print(x // 10)
print(x // z)
```

#### Modulus (`%`)
- Modulus returns the remainder after division.
- It is commonly used to check divisibility and even or odd numbers.

```python
x = 10
z = 20

print(x % 10)
print(x % z)
print(10 % 3)
```

#### Exponentiation (`**`)
- Exponentiation is used to calculate powers.
- It can also be used with fractional powers, such as square roots.

```python
print(2 ** 4)
print(5 ** 5)
print(25 ** 0.5)
```

### Operator Precedence in Python
When multiple arithmetic operators are used in one expression, Python follows an order of precedence:

1. Parentheses `()` are evaluated first. Example: `(2 + 3) * 4 = 20`
2. Exponentiation `**` is evaluated next. Example: `3 ** 2 = 9`
3. Multiplication `*`, division `/`, floor division `//`, and modulus `%` are evaluated from left to right. Example: `10 / 2 * 3 = 15`

## Assignment Operators
- Assignment operators are used to store values in variables.
- The basic assignment operator is `=`.
- Compound assignment operators include `+=`, `-=`, `*=`, `/=`, `//=`, `%=` and `**=`.
- Python evaluates the expression first and then assigns the result to the variable.

### Basic Assignment (`=`)
- Use `=` to store a value in a variable.
- Once assigned, the variable can be reused later in the program.

```python
x=10
print(x)
```

### Assigning `None`
- `None` represents the absence of a value.
- It is often used as a placeholder until a real value is assigned.

```python
a=None
print(a,type(a))
```

### Valid Expression Before Assignment
- Python first evaluates the expression on the right-hand side.
- If the expression is invalid, Python raises an error instead of completing the assignment.

```python
b=int("10.4")
```

### Compound Assignment
- Compound assignment gives a shorter way to update the same variable.
- For example, `c += 10` means `c = c + 10`.

#### Standard Update with `=`
- This is the longer way to change a variable by writing the full expression again.

```python
c=50

c= c+10

print(c)  # these are normal 

c= c-30

print(c)
```

#### Compound Assignment Shortcut
- This shorter form updates the same variable in one step.

```python
c=50

c+=10

print(c) # these are compund assignment

c-=30

print(c)
```

## Comparison Operators
- Comparison operators are used to compare two values.
- The result of a comparison is always `True` or `False`.
- Common comparison operators are `==`, `!=`, `<`, `>`, `<=`, and `>=`.

### Equal To (`==`)
- `==` checks whether two values are equal.
- If both values are the same, Python returns `True`; otherwise it returns `False`.

```python
a=10
b=10.0

print(10==10)
print(10==10.0)
print(type(10)==type(10.0))
print(True==bool(10.4))
print('Abhi'=='abhi')
```

### Not Equal To (`!=`)
- `!=` checks whether two values are different.
- If the values are not the same, Python returns `True`.

```python
print(a!=b)
print(10.1!=a)
```

### Greater Than (`>`) and Less Than (`<`)
- `>` checks whether the value on the left is greater than the value on the right.
- `<` checks whether the value on the left is less than the value on the right.
- Both operators return `True` or `False`.

```python
print(100>10)
print(140>100)
print(100<10)
print(100<140)
```

### Less Than or Equal To (`<=`)
- `<=` checks whether the value on the left is less than or equal to the value on the right.
- It returns `True` when both values are equal or when the left value is smaller.

```python
print(10 <= 10)
print(8 <= 10)
print(12 <= 10)
```

### Greater Than or Equal To (`>=`)
- `>=` checks whether the value on the left is greater than or equal to the value on the right.
- It returns `True` when both values are equal or when the left value is larger.

```python
print(10 >= 10)
print(15 >= 10)
print(5 >= 10)
```

### Chained Comparison Operators
- Python allows multiple comparisons in one expression, such as `10 < x < 20`.
- This checks whether the value satisfies all conditions at the same time.
- The result is `True` only when every comparison in the chain is true.

```python
x = 15

print(10 < x < 20)
print(10 <= x <= 15)
print(5 > x > 1)
print(1 < x >= 15)
```

## Logical Operators
Logical operators are used to combine or change conditions in Python.

- `and` operator: Returns `True` only when all given conditions are `True`.  
- `or` operator: Returns `True` if at least one condition is `True`.
- `not` operator: Reverses the result, changing `True` to `False` and `False` to `True`.

### `and` Operator

```python
print(True and True)
print(False and True)
print(True and False)
print(False and False)
```

```python
print('Kumod'=='Kumod' and 'and'=='and')
```

```python
print(type(int(56.00)) and type(float("nan")))
```

#### Why does this print `<class 'float'>`?

- `int(56.00)` becomes `56`, so `type(int(56.00))` is `<class 'int'>`.
- `float("nan")` creates a special floating-point value called `nan`, so `type(float("nan"))` is `<class 'float'>`.
- In Python, `and` does not always return `True` or `False`.
- It returns the first falsy value, or if both values are truthy, it returns the second value.
- `<class 'int'>` is truthy, so Python returns `<class 'float'>`.

### `or` Operator

```python
print(True or True)
print(False or True)
print(False or False)
print(0 or "Python")
```

#### How does `or` work?

- `or` returns `True` if at least one condition is `True`.
- If both conditions are `False`, the result is `False`.
- With non-boolean values, Python returns the first truthy value.
- In `0 or "Python"`, `0` is falsy and `"Python"` is truthy, so Python returns `"Python"`.

### `not` Operator

```python
print(not True)
print(not False)

x = 10
print(not (x < 5))
```

#### How does `not` work?

- `not` reverses the result of a condition.
- `not True` becomes `False`.
- `not False` becomes `True`.
- Here, `x < 5` is `False` because `x` is `10`, so `not (x < 5)` becomes `True`.

## Membership Operators

Membership operators are used to check whether a value exists inside a sequence like a string, list, or tuple.

- `in` operator: Returns `True` if the value is present.
- `not in` operator: Returns `True` if the value is not present.

### `in` and `not in` Operators

```python
name = "Python"
numbers = [10, 20, 30, 40]

print("Py" in name)
print("java" in name)
print(20 in numbers)
print(100 not in numbers)
```

#### How do membership operators work?

- `"Py" in name` is `True` because `"Py"` is present in `"Python"`.
- `"java" in name` is `False` because that text is not present.
- `20 in numbers` is `True` because `20` exists in the list.
- `100 not in numbers` is `True` because `100` is not in the list.

## Identity Operators

Identity operators are used to check whether two variables refer to the same object in memory.

- `is` operator: Returns `True` if both variables point to the same object.
- `is not` operator: Returns `True` if both variables point to different objects.

### `is` and `is not` Operators

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

d="A"
e='a'



print(e>d)
#int(d)
print(type(d))
print(a is b)
print(a is c)
print(a == c)
print(a is not c)
```

#### How do identity operators work?

- `a is b` is `True` because `b` refers to the same list object as `a`.
- `a is c` is `False` because `c` is a different list object.
- `a == c` is `True` because both lists have the same values.
- `a is not c` is `True` because `a` and `c` are different objects in memory.
