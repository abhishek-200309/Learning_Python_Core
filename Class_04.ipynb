- Symbol kind of operators => +,-,*,/,% >,<,!=,>=,<=,==
- Keywords kind of operators => and ,or , not ,in , not in

## Tyeps of Operators
- Arithmetic 
- Assignment 
- Comparison

### Arithmetic Operators

```python
x=10
a=15

print(x+a)

print(a-x)

print(x*a)

print(a/x)


```

### Floor Division/double division/integer division
- It is widely used to build dynamic codes.
- it returns quotient value before going to the decimal places.

```python


x=10
z=20

print(x//10)

print(x//z)
```

### Modulus => 
- (%) it returns the remainder value.
- it returns the remainder value, before going to decimal places.

```python
x=10
z=20

print(x%10)

print(x%z)

print(10%3)
```

### Exponentation => power(**)
-

```python
print(2**4)
print(5**5)
print(25**.5)
```

### Multiple Arithmetic Operators Usage.

- Order Precedence in Python
1. Praentheses() => Operations inside parentheses are evaluated first. Ex=> (2+3)*4= 5*4=20
2. Exponentiation => Powers are calculated after parentheses.
Ex=> 3**2=9
3. Multiplication (*),division(/),floor division (//),modulus(%)
=> All four have the same precedence. if multiple appear. Python evaluates them left to right.
Ex=> 10/2*3 = 5*3=15
